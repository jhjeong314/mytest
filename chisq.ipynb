```r
alpha = 0.05
n = 15
s2 = 0.20
l = (n-1)/qchisq(1-alpha/2, n-1) # left coeff. 
r = (n-1)/qchisq(alpha/2, n-1) # right coeff.
cat(sprintf("Conf. Interval = (%.3f, %.3f)\n", s2*l, s2*r))
```

```r
alpha = 0.05
nx = 13; ny = 11
sdevx = 2.3; sdevy = 1.5
l = 1/qf(1-alpha/2, nx-1, ny-1) # left coeff. 
r = 1/qf(alpha/2, nx-1, ny-1) # left coeff. 
cat(sprintf("Conf. Interval = (%.3f, %.3f)\n", sdevx/sdevy*l, sdevx/sdevy*r))
```

```r
x = seq(0,21,length=200)
ychi2 = dchisq(x,5)
plot(x,ychi2,type="l")
abline(h=0, col="#aaaaaa")
```

```r
qchisq(0.025,5)
qchisq(0.975,5)
```

```r
xchi2 = x
for(i in 1:200) {
    cat(sprintf("{%.3f,%.4f},",xchi2[i],ychi2[i]))
    if(i%%5 == 0) cat("\n")
}
```
